```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
BeautifulSoup('test','lxml')
```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
title = soup.title
print(title)
```

```python
text = soup.get_text()
```

```python
rows = soup.find_all('tr')
print(rows[:10])
```

```python
for row in rows:
    row_td = row.find_all('td')
print(row_td)
type(row_td)
```

```python
list_rows = []
for row in rows:
    row_td = row.find_all('td')
    str_cells = str(row_td)
    cleantext = BeautifulSoup(str_cells,"lxml").get_text()
    list_rows.append(cleantext)
    print(cleantext)
    type(cleantext)
```

```python
import re
list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '', str_cells))
    list_rows.append(clean2)
    print(clean2)
    type(clean2)
```

```python
Postal_Codes = pd.DataFrame(list_rows)
Postal_Codes.head(10)
```

```python
Postalcodes = Postal_Codes[0].str.split(',', expand = True)
Postalcodes.head(10)
```

```python
df = Postalcodes.drop(Postalcodes.columns[3:31], axis=1)
df.head(10)
```

```python
df[0] = df[0].str.strip('[\n')
df.head(10)
```

```python
df[1] = df[1].str.strip('\n')
df.head(10)
```

```python
col_labels = soup.find_all('th')
```

```python
all_header = []
col_str = str(col_labels)
cleantext2 = BeautifulSoup(col_str, "lxml").get_text()
all_header.append(cleantext2)
print(all_header)
```

```python
PCHeaders = pd.DataFrame(all_header)
PCHeaders.head()
```

```python
PCHeaders = PCHeaders[0].str.split(',', expand = True)
PCHeaders.head()
```

```python
PCHeaders = PCHeaders.drop(PCHeaders.columns[3], axis=1)
PCHeaders.head()
```

```python
PCHeaders[0] = PCHeaders[0].str.strip('[\n')
PCHeaders.head(10)
```

```python
PCHeaders[1] = PCHeaders[1].str.strip('\n')
PCHeaders.head(10)
```

```python
PCHeaders[2] = PCHeaders[2].str.strip('\n')
PCHeaders.head(10)
```

```python
frames = [PCHeaders, df]

Postal_Codes_Canada = pd.concat(frames)
Postal_Codes_Canada.head(10)
```

```python
Postal_Canada = Postal_Codes_Canada.rename(columns=Postal_Codes_Canada.iloc[0])
Postal_Canada.head(10)
```
